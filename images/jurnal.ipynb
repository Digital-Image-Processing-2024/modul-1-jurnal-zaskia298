```python
import matplotlib.pyplot as plt
import numpy as np

# Baca citra grayscale
gray_image = plt.imread("aang.jpg")
gray_image = np.dot(gray_image[..., :3], [0.2989, 0.5870, 0.1140])

tinggi, lebar = gray_image.shape

print("Tinggi:", tinggi)
print("Lebar:", lebar)

# Hitung histogram
hist, bins = np.histogram(gray_image, bins=250, range=(0, 1))

hist_normalized = hist / np.sum(hist)

# Plot histogram
plt.bar(bins[:1], hist, color="green", width=1)
plt.ylabel("Normalize Frequency")
plt.show()
```
